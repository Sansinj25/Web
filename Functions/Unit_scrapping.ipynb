```python
import numpy as np
import pandas as pd
import re
import nltk
import itertools
from nltk.corpus import wordnet as wn
from nltk import pos_tag
from nltk.tag import UnigramTagger
from nltk.corpus import brown
import requests
from bs4 import BeautifulSoup
import string
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
ingredient=[]
ingred_unities=[]

categories=['Vegetables+and+Vegetable+Products','Cereal+Grains+and+Pasta','Dairy+and+Egg+Products',\
            'Poultry+Products','Fats+and+Oils','Spices+and+Herbs','Fruits+and+Fruit+Juices',\
            'Legumes+and+Legume+Products','Nut+and+Seed+Products','Pork+Products','Beef+Products',\
            'Lamb%2C+Veal%2C+and+Game+Products','Sweets','Baked+Products']
page_numbers={categories[0]:35,categories[1]:19,categories[2]:25,categories[3]:25,categories[4]:19,\
             categories[5]:6,categories[6]:25,categories[7]:25,categories[8]:10,categories[9]:4,\
              categories[10]:4,categories[11]:4,categories[12]:6,categories[13]:3}

for cat in categories:
    print(cat)
    sum_ingr_added=0
    for i in range(1,page_numbers[cat]+1):
        url='https://www.nutritionvalue.org/search.php?category='+cat+'&page='+str(i)
        r=requests.get(url,headers=headers)
        page_body = r.text
        soup=BeautifulSoup(page_body, 'html.parser')
        stat= soup.find('table', class_="full_width results zero")
        ingr_added=0
        for ingred in stat.find_all('td',class_="results left"):
            a=ingred.find('a')
            ingr_name=a.get('title')
            ingredient.append(ingr_name)
            url2='https://www.nutritionvalue.org'+a.get('href')
            r=requests.get(url2,headers=headers)
            page_body = r.text
            soup=BeautifulSoup(page_body, 'html.parser')
            selec=soup.find('select')
            if selec:
                ingr_added+=1
            dic_values={}
            for q in selec.find_all('option'):
                ratios_values=q.get('value').split('=')
                if len(ratios_values)>1:
                    dic_values[ratios_values[0]]=ratios_values[1]
                else:
                    dic_values[ratios_values[0]]=np.nan
            ingred_unities.append(dic_values)
        print('ingredients ajoutés:',ingr_added)
        sum_ingr_added+=ingr_added
    print('Total ingredients in: ' ,cat,'is :',sum_ingr_added)
```

```python
df=pd.DataFrame(data=ingred_unities,index=ingredient)
display(df.head())
```

```python
path='../ADA_JEX2017/Project/Functions/'
with open(path+'ingredient_list.txt', 'r') as f:
    ingredient_list = [line.rstrip('\n') for line in f]
    
lemmatizer = WordNetLemmatizer()
ingredient_list=[lemmatizer.lemmatize(token).lower() for token in ingredient_list]
```

```python
#      option 1
##  TAKE ELEM THAT WE ALREADY HAVE:
categ_list=[Vegi,Cereal,Dairy,Poultry,Fats,Spices,Fruits,Legumes,Nuts,Pork,Beef,Sweet,Lamb,Baked]
match_ingred=[]
unmatch=[]
matching_dic={}
for cat in categ_list:
    for i in cat:
        matched=[]
        for elem in i.split(', '):
            elem=lemmatizer.lemmatize(elem.lower())
            if elem in ingredient_list and (elem not in match_ingred):
                match_ingred.append(elem)
                matched.append(elem)
            elif (elem not in unmatch):
                unmatch.append(elem)
        if matched:
            matching_dic[i.lower()]=matched[0]
```
